# **M√©tricas de Clustering**

El clustering es una t√©cnica de aprendizaje no supervisado que agrupa un conjunto de datos en grupos o "clusters" bas√°ndose en caracter√≠sticas comunes. Los datos dentro de un mismo grupo son m√°s similares entre s√≠ que con los de otros grupos.

Ejemplos comunes de aplicaciones de clustering incluyen segmentaci√≥n de clientes, agrupaci√≥n de im√°genes, y detecci√≥n de anomal√≠as.

Diferencia entre Clustering Jer√°rquico y No Jer√°rquico:

Clustering Jer√°rquico:

Crea una jerarqu√≠a de clusters. Puede ser aglomerativo (comienza con puntos individuales y fusiona) o divisivo (comienza con un solo cluster y divide).
Produce un dendrograma que ilustra la relaci√≥n entre los clusters.

Clustering No Jer√°rquico:

Asigna los datos a un n√∫mero fijo de clusters (ej. K-means).
Generalmente m√°s r√°pido y m√°s f√°cil de implementar, pero no proporciona informaci√≥n sobre la jerarqu√≠a de los clusters.

**¬øPor qu√© son importantes las m√©tricas de clustering?**

**Evaluaci√≥n del Desempe√±o del Modelo:**

Las m√©tricas permiten evaluar cu√°n bien el algoritmo ha agrupado los datos. Sin una m√©trica adecuada, es dif√≠cil determinar si el modelo est√° funcionando correctamente.
Las m√©tricas ayudan a identificar si los clusters son coherentes y √∫tiles para el an√°lisis posterior.

**Comparaci√≥n entre Diferentes Algoritmos de Clustering:**

Al aplicar diferentes m√©todos de clustering, es esencial utilizar m√©tricas para comparar su desempe√±o.

Por ejemplo, al evaluar los resultados de K-means frente a DBSCAN, las m√©tricas pueden mostrar cu√°l modelo produce grupos m√°s cohesivos y separados.

**Tipos de M√©tricas**

**M√©tricas para Problemas de Clustering:**

**M√©tricas Internas:** Se calculan a partir de la misma agrupaci√≥n (ej. Cohesi√≥n, Inercia, √çndice de Silueta).

**Cohesi√≥n:** Mide cu√°n cerca est√°n los puntos dentro del mismo cluster.

**Inercia**: Mide la suma de las distancias al cuadrado entre los puntos y el centroide de su cluster.

**√çndice de Silueta**: Mide qu√© tan bien se separan los clusters.

**M√©tricas Externas**: Comparan los resultados de clustering con un conjunto de referencia o etiquetas conocidas (ej. Rand Index, Jaccard Index).

**Rand Index**: Mide la similaridad entre dos agrupaciones.

**Jaccard Index**: Mide la similaridad entre conjuntos de datos.

**M√©tricas de Estabilidad:** Evaluaci√≥n de la consistencia de los clusters al aplicar diferentes inicializaciones o subconjuntos de datos.

**El m√©todo de codo (o "Elbow Method")**

Es una t√©cnica com√∫nmente utilizada para determinar el n√∫mero √≥ptimo de clusters en un algoritmo de clustering, como K-means. Este m√©todo se basa en la inercia, que mide la compactaci√≥n de los clusters

**¬øC√≥mo funciona el m√©todo de codo?**

Se ejecuta el algoritmo de clustering (por ejemplo, K-means) para diferentes valores de
k (el n√∫mero de clusters).

Para cada k, se calcula la inercia total (la suma de las distancias cuadradas entre los puntos y sus centroides).

Se grafica la inercia en funci√≥n dek. En el eje x se coloca el n√∫mero de clusters k, y en el eje y la inercia total.

A medida que el n√∫mero de clusters aumenta, la inercia disminuye (porque los puntos est√°n m√°s cerca de los centroides), pero la mejora empieza a disminuir a partir de cierto punto.

El "codo" del gr√°fico es el punto donde la tasa de disminuci√≥n de la inercia se ralentiza significativamente. Ese punto es donde se considera que est√° el n√∫mero √≥ptimo de clusters, ya que aumentar m√°s el n√∫mero de clusters no mejora sustancialmente la compactaci√≥n de los puntos.

En la parte inicial del gr√°fico, la inercia disminuye r√°pidamente cuando agregas m√°s clusters.
Despu√©s de cierto punto, el beneficio de agregar m√°s clusters disminuye y la curva empieza a "suavizarse", creando un √°ngulo en forma de codo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos para los clusters
cluster_a = np.array([[1, 2], [2, 1], [1, 1]])
cluster_b = np.array([[5, 6], [6, 5], [5, 5]])
cluster_c = np.array([[9, 8], [8, 9], [9, 9]])
```

```python
# Graficar solo los datos
plt.figure(figsize=(8, 6))

plt.scatter(cluster_a[:, 0], cluster_a[:, 1])
plt.scatter(cluster_b[:, 0], cluster_b[:, 1])
plt.scatter(cluster_c[:, 0], cluster_c[:, 1])

# Configuraci√≥n del gr√°fico
plt.title('Datos')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calcular centroides
centroid_a = cluster_a.mean(axis=0)
centroid_b = cluster_b.mean(axis=0)
centroid_c = cluster_c.mean(axis=0)
```

```python
# Funci√≥n para calcular la inercia
def calcular_inercia(cluster, centroide):
    distancias = np.sum((cluster - centroide) ** 2, axis=1)
    return np.sum(distancias)
```

```python
# Inercia para cada cluster
inercia_a = calcular_inercia(cluster_a, centroid_a)
inercia_b = calcular_inercia(cluster_b, centroid_b)
inercia_c = calcular_inercia(cluster_c, centroid_c)
inercia_total = inercia_a + inercia_b + inercia_c

# Mostrar resultados
print(f"Inercia A: {inercia_a:.2f}")
print(f"Inercia B: {inercia_b:.2f}")
print(f"Inercia C: {inercia_c:.2f}")
print(f"Inercia Total: {inercia_total:.2f}")
```

```python
from sklearn.cluster import KMeans

# Datos combinados para clustering
data = np.vstack((cluster_a, cluster_b, cluster_c))

# Lista para guardar las inercias
inercias = []

# Rango de k (n√∫mero de clusters)
k_values = range(1, 10)

# Calcular inercia para cada k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    inercias.append(kmeans.inertia_)

# Graficar el m√©todo del codo
plt.figure(figsize=(8, 6))
plt.plot(k_values, inercias, marker='o', linestyle='-', color='b')
plt.title('M√©todo del Codo')
plt.xlabel('N√∫mero de Clusters (k)')
plt.ylabel('Inercia Total')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
# Visualizaci√≥n
clusters = [cluster_a, cluster_b, cluster_c]
centroides = [centroid_a, centroid_b, centroid_c]
colores = ['red', 'blue', 'green']

plt.figure(figsize=(8, 6))

for cluster, centroide, color in zip(clusters, centroides, colores):
    plt.scatter(cluster[:, 0], cluster[:, 1], color=color, label=f'Cluster {color.capitalize()}')
    plt.scatter(*centroide, color=color, edgecolor='black', s=150, marker='X', label=f'Centroide {color.capitalize()}')
    for punto in cluster:
        plt.plot([punto[0], centroide[0]], [punto[1], centroide[1]], color=color, linestyle='--', alpha=0.7)

plt.title('Visualizaci√≥n de Clusters e Inercia')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid(True)
plt.show()
```

**Valores cercanos a 1**: Un √≠ndice de silueta alto (cercano a 1) indica que el punto i est√° bien agrupado, es decir, m√°s cerca de los puntos de su propio cluster que de los puntos en otros clusters. Esto significa que el clustering es de buena calidad.

**Valores cercanos a 0:** Un √≠ndice de silueta cerca de 0 sugiere que el punto i est√° en la frontera entre dos clusters, lo que indica que no est√° claro a qu√© cluster pertenece.

**Valores negativos:** Un √≠ndice de silueta negativo indica que el punto est√° m√°s cerca de los puntos en el cluster m√°s cercano que de su propio cluster, lo que sugiere que el clustering es deficiente y el punto podr√≠a estar mal clasificado.

## Cluster A

**Puntos:**
- (1, 2)

$$
c_A = \left(\frac{1+2+1}{3}, \frac{2+1+1}{3}\right) = (1.33, 1.33)

### C√°lculo de \(a(i)\) (distancia media dentro del mismo cluster):

- Para el punto (1, 2):
$$
a(1,2) = \frac{\| (1,2) - (2,1) \| + \| (1,2) - (1,1) \|}{2} = \frac{ \sqrt{(1-2)^2 + (2-1)^2} + \sqrt{(1-1)^2 + (2-1)^2}}{2} = 1.21
$$

- Para el punto (2, 1):
$$
a(2,1) = \frac{\| (2,1) - (1,2) \| + \| (2,1) - (1,1) \|}{2} = \frac{ \sqrt{(2-1)^2 + (1-2)^2} + \sqrt{(2-1)^2 + (1-1)^2}}{2} = 1.21
$$

- Para el punto (1, 1):
$$
a(1,1) = \frac{\| (1,1) - (1,2) \| + \| (1,1) - (2,1) \|}{2} = \frac{ \sqrt{(1-1)^2 + (1-2)^2} + \sqrt{(1-2)^2 + (1-1)^2}}{2} = 1.0
$$

### C√°lculo de \(b(i)\) (distancia media al cluster m√°s cercano):

Supongamos que el cluster m√°s cercano es el Cluster B con centroide \(c_B = (5.33, 5.33)\).

- Para el punto (1, 2):
$$
b(1,2) = \| (1,2) - (5.33, 5.33) \| = \sqrt{(1-5.33)^2 + (2-5.33)^2} = 5.23
$$

- Para el punto (2, 1):
$$
b(2,1) = \| (2,1) - (5.33, 5.33) \| = \sqrt{(2-5.33)^2 + (1-5.33)^2} = 5.78

- Para el punto (1, 1):
$$
b(1,1) = \| (1,1) - (5.33, 5.33) \| = \sqrt{(1-5.33)^2 + (1-5.33)^2} = 6.13
$$

### C√°lculo del √çndice de Silueta (\(s(i)\)):

El √≠ndice de silueta para un punto se calcula como:

$$
s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}
$$

- Para el punto (1, 2):
$$
s(1,2) = \frac{5.23 - 1.21}{\max(1.21, 5.23)} = \frac{4.02}{5.23} = 0.77
$$

- Para el punto (2, 1):
$$
s(2,1) = \frac{5.78 - 1.21}{\max(1.21, 5.78)} = \frac{4.57}{5.78} = 0.79
$$

- Para el punto (1, 1):
$$
s(1,1) = \frac{6.13 - 1.0}{\max(1.0, 6.13)} = \frac{5.13}{6.13} = 0.84
$$

**√çndice de Silueta Promedio para Cluster A:**
$$
s_A = \frac{s(1,2) + s(2,1) + s(1,1)}{3} = \frac{0.77 + 0.79 + 0.84}{3} = 0.80

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm
```

```python
# Datos de los puntos en los clusters
X = np.array([
    [1, 2], [2, 1], [1, 1],  # Cluster A
    [5, 6], [6, 5], [5, 5],  # Cluster B
    [8, 1], [9, 2], [8, 2]   # Cluster C
])
```

```python
# Probar diferentes n√∫meros de clusters
for n_clusters in range(2, 5):
    # Crear un subplot de 1 fila y 2 columnas
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # Configuraci√≥n del gr√°fico de silueta
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Ajustar el modelo KMeans
    clusterer = KMeans(n_clusters=n_clusters, max_iter=1000, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # Calcular el promedio del √≠ndice de silueta
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"Para n_clusters = {n_clusters}, el promedio del score de silhouette es: {silhouette_avg}")

    # Calcular el √≠ndice de silueta para cada punto
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Obtener los valores de silueta para el cluster i
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Etiqueta de cada cluster
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        y_lower = y_upper + 10  # Ajustar y_lower para el siguiente cluster

    ax1.set_title("Gr√°fico de Silueta para varios clusters")
    ax1.set_xlabel("Coeficientes de Silueta")
    ax1.set_ylabel("Cluster")
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax1.set_yticks([])
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # Visualizaci√≥n de los clusters
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=100, lw=0, alpha=0.7, c=colors, edgecolor='k')

    # Dibujar centroides
    centers = clusterer.cluster_centers_
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o', c="white", alpha=1, s=200, edgecolor='k')
    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker=f'$%d$' % i, alpha=1, s=50, edgecolor='k')

    ax2.set_title("Visualizaci√≥n de Clusters")
    ax2.set_xlabel("Variable X")
    ax2.set_ylabel("Variable Y")

    plt.suptitle((f"An√°lisis de Silueta para KMeans con n_clusters = {n_clusters}"),
                 fontsize=14, fontweight='bold')

    plt.show()
```

##**√çndice de Rand (RI)**

El √çndice de Rand (RI) es una m√©trica utilizada para evaluar la similitud entre dos particiones de un conjunto de datos, como una partici√≥n verdadera (etiquetas reales) y una partici√≥n obtenida por un algoritmo de clustering. Mide qu√© tan bien los objetos han sido agrupados en comparaci√≥n con una referencia.

La versi√≥n ajustada del √çndice de Rand (ARI) corrige el valor esperado del RI bajo una asignaci√≥n aleatoria de etiquetas y lo normaliza para garantizar que el valor est√© entre
‚àí
1 y
1.

El RI compara todas las posibles combinaciones de pares de puntos en el dataset y calcula la proporci√≥n de pares que est√°n consistentemente etiquetados en ambas particiones.

La f√≥rmula del RI es:

$$
RI = \frac{TP + TN}{\text{Total de Pares}}
$$


Donde:

TP (True Positives): Pares de puntos que est√°n en el mismo cluster en ambas particiones.

TN (True Negatives): Pares de puntos que est√°n en diferentes clusters en ambas particiones.


Total de pares

FP (False Positives): Pares de puntos que est√°n en el mismo cluster en la partici√≥n predicha, pero en diferentes clusters en la referencia.

FN (False Negatives): Pares de puntos que est√°n en diferentes clusters en la predicci√≥n, pero en el mismo cluster en la referencia.


El ARI toma en cuenta que el valor esperado del RI puede ser alto incluso para asignaciones aleatorias. Valores cercanos a
1 indican un buen clustering, mientras que valores cercanos a
0 o negativos indican bajo acuerdo con la referencia.

**Clustering generado:**
- **Cluster A**: \( (1,2), (2,1), (1,1) \)
- **Cluster B**: \( (5,6), (6,5), (5,5) \)
- **Cluster C**: \( (8,9), (9,8), (8,8) \)

**Clustering real (verdad ya conocida):**

- **Cluster 1**: \( (1,2), (2,1), (1,1) \)
- **Cluster 2**: \( (5,6), (6,5), (5,5) \)
- **Cluster 3**: \( (8,9), (9,8), (8,8) \)

Primero calculamos los pares consistentes e inconsistentes:

**Total de pares posibles:**  
Para \( n = 9 \) puntos:

$$
\binom{9}{2} = \frac{9 \cdot 8}{2} = 36
$$

**Pares consistentes (TP + TN):**

Todos los puntos est√°n agrupados exactamente igual en ambas particiones.

$$
TP = 3 \cdot \binom{3}{2} = 3 \cdot 3 = 9 \quad \text{(pares dentro de los clusters correctos)}
$$

$$
TN = 36 - 9 = 27 \quad \text{(pares entre diferentes clusters)}
$$

**C√°lculo del RI:**

$$
RI = \frac{TP + TN}{\text{Total de Pares}} = \frac{9 + 27}{36} = 1
$$


**√çndice de Rand Ajustado (ARI)**

El ARI ajusta el RI restando el valor esperado y normaliz√°ndolo:

$$
ARI = \frac{RI - \text{Valor Esperado}}{\text{M√°ximo RI} - \text{Valor Esperado}}
$$

```python
import numpy as np
from sklearn.metrics import adjusted_rand_score
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Puntos originales de los clusters y sus etiquetas verdaderas
cluster_a = [(1, 2), (2, 1), (1, 1)]
cluster_b = [(5, 6), (6, 5), (5, 5)]
cluster_c = [(8, 1), (9, 2), (9, 1)]

# Concatenar todos los puntos
X = np.array(cluster_a + cluster_b + cluster_c)
true_labels = [0] * len(cluster_a) + [1] * len(cluster_b) + [2] * len(cluster_c)

# Realizar clustering con KMeans
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
predicted_labels = kmeans.fit_predict(X)

# Calcular el √çndice de Rand Ajustado
ari = adjusted_rand_score(true_labels, predicted_labels)
print(f"√çndice de Rand Ajustado (ARI): {ari:.4f}")

# Graficar los clusters reales y los predichos
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Visualizaci√≥n de los clusters reales
axes[0].scatter(X[:, 0], X[:, 1], c=true_labels, cmap=cm.nipy_spectral, s=100, edgecolor='k')
axes[0].set_title("Clusters Reales", fontsize=14)
axes[0].set_xlabel("X1")
axes[0].set_ylabel("X2")

# Visualizaci√≥n de los clusters predichos
axes[1].scatter(X[:, 0], X[:, 1], c=predicted_labels, cmap=cm.nipy_spectral, s=100, edgecolor='k')
axes[1].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                c='white', s=300, edgecolor='k', label='Centroides')
for i, center in enumerate(kmeans.cluster_centers_):
    axes[1].scatter(center[0], center[1], marker=f'$ {i} $', s=200, c='black')
axes[1].set_title("Clusters Predichos", fontsize=14)
axes[1].set_xlabel("X1")
axes[1].set_ylabel("X2")

plt.suptitle("Comparaci√≥n de Clusters Reales vs Predichos", fontsize=16)
plt.legend()
plt.tight_layout()
plt.show()
```

**Clusters reales:** Muestra la distribuci√≥n seg√∫n las etiquetas originales.

**Clusters predichos:** Visualiza los resultados del KMeans y sus centroides.

## **√çndice de Calinski-Harabasz**

El √çndice de Calinski-Harabasz (tambi√©n conocido como el √çndice de la Varianza Entre los Cl√∫steres o √çndice de Variabilidad de la Clase), es una m√©trica utilizada para evaluar la calidad de una partici√≥n de cl√∫steres en un conjunto de datos. Este √≠ndice compara la dispersi√≥n entre los cl√∫steres con la dispersi√≥n dentro de los cl√∫steres, lo que permite evaluar qu√© tan bien definidos est√°n los cl√∫steres.

# C√°lculo del √çndice de Calinski-Harabasz (CH)

El √çndice de Calinski-Harabasz (CH) se utiliza para evaluar la calidad de un clustering. La f√≥rmula general es la siguiente:

$$
CH = \frac{Tr(B_k)}{Tr(W_k)} \cdot \frac{n - k}{k - 1}
$$

Donde:

- \( Tr(B_k) \): La traza de la matriz de dispersi√≥n entre los cl√∫steres. Representa la dispersi√≥n entre los centros de los cl√∫steres.
- \( Tr(W_k) \): La traza de la matriz de dispersi√≥n dentro de los cl√∫steres. Representa la dispersi√≥n dentro de los puntos de cada cl√∫ster.
- \( n \): N√∫mero total de puntos en el dataset.
- \( k \): N√∫mero de cl√∫steres en la partici√≥n.

### Cl√∫steres utilizados en el ejemplo

#### Clustering generado:
- **Cluster A**: \( (1,2), (2,1), (1,1) \)
- **Cluster B**: \( (5,6), (6,5), (5,5) \)
- **Cluster C**: \( (8,9), (9,8), (8,8) \)

#### Clustering real (verdad conocida):
- **Cluster 1**: \( (1,2), (2,1), (1,1) \)
- **Cluster 2**: \( (5,6), (6,5), (5,5) \)
- **Cluster 3**: \( (8,9), (9,8), (8,8) \)

### Paso 1: Calcular \( Tr(B_k) \) (Dispersi√≥n entre los cl√∫steres)

La dispersi√≥n entre los cl√∫steres se calcula como la varianza de los centros de los cl√∫steres con respecto al centro global (centroide de todos los puntos).

#### Centro de cada cl√∫ster:
- **Centro A**:
  $$
  A = \left( \frac{1+2+1}{3}, \frac{2+1+1}{3} \right) = (1.33, 1.33)
  $$
- **Centro B**:
  $$
  B = \left( \frac{5+6+5}{3}, \frac{6+5+5}{3} \right) = (5.33, 5.33)
  $$
- **Centro C**:
  $$
  C = \left( \frac{8+9+8}{3}, \frac{9+8+8}{3} \right) = (8.33, 8.33)
  $$

#### Centro global del dataset:

El centro global es el centroide de todos los puntos en el conjunto de datos:

$$
Centro\ Global = \left( \frac{1+2+1+5+6+5+8+9+8}{9}, \frac{2+1+1+6+5+5+9+8+8}{9} \right) = (5.33, 5.33)
$$

#### C√°lculo de \( Tr(B_k) \):

La dispersi√≥n entre los cl√∫steres \( Tr(B_k) \) es la suma de las distancias cuadradas entre el centro de cada cl√∫ster y el centro global:

$$
Tr(B_k) = \sum_{i=1}^{k} n_i \cdot (Distancia\ entre\ centro\ i\ y\ centro\ global)^2
$$

Donde \( n_i \) es el n√∫mero de puntos en el cl√∫ster \( i \).

### Paso 2: Calcular \( Tr(W_k) \) (Dispersi√≥n dentro de los cl√∫steres)

La dispersi√≥n dentro de los cl√∫steres \( Tr(W_k) \) se calcula como la varianza dentro de cada cl√∫ster, es decir, la suma de las distancias cuadradas entre cada punto del cl√∫ster y su centro.

$$
Tr(W_k) = \sum_{i=1}^{k} \sum_{x_j \in C_i} (Distancia\ entre\ x_j\ y\ centro\ i)^2
$$

### Paso 3: Calcular el √çndice de Calinski-Harabasz

Una vez que tengamos \( Tr(B_k) \) y \( Tr(W_k) \), podemos sustituir estos valores en la f√≥rmula original:

$$
CH = \frac{Tr(B_k)}{Tr(W_k)} \cdot \frac{n - k}{k - 1}
$$

Donde:

- \( n = 9 \) (el n√∫mero total de puntos).
- \( k = 3 \) (el n√∫mero de cl√∫steres).

Este √≠ndice es mayor cuando los cl√∫steres est√°n mejor definidos, es decir, cuando las distancias dentro de los cl√∫steres son peque√±as y las distancias entre los cl√∫steres son grandes.

### Interpretaci√≥n del √çndice

- Un valor **alto** del √≠ndice sugiere que los cl√∫steres son compactos y bien separados, lo que indica una buena partici√≥n.
- Un valor **bajo** sugiere que los cl√∫steres est√°n mal definidos y pueden estar solapados.
  
Este √≠ndice es √∫til para evaluar el rendimiento de un algoritmo de clustering, como K-means, sin necesidad de tener etiquetas de clase reales (es decir, no requiere una "verdad conocida").


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
```

```python
# Definir los puntos de los cl√∫steres
cluster_a = np.array([[1, 2], [2, 1], [1, 1]])
cluster_b = np.array([[5, 6], [6, 5], [5, 5]])
cluster_c = np.array([[8, 9], [9, 8], [8, 8]])

# Unir todos los puntos en un √∫nico array
X = np.vstack([cluster_a, cluster_b, cluster_c])

# Realizar el clustering utilizando KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Calcular el √çndice de Calinski-Harabasz
ch_score = calinski_harabasz_score(X, kmeans.labels_)

# Mostrar el √≠ndice
print(f"√çndice de Calinski-Harabasz: {ch_score}")
```

## **√çndice de Davies-Bouldin (DBI)**

El √çndice de Davies-Bouldin (DBI) es una m√©trica utilizada para evaluar la calidad de un clustering. Su objetivo es medir la separaci√≥n y compacidad de los cl√∫steres obtenidos mediante un algoritmo de clustering, como K-means. Un valor bajo del √≠ndice indica un buen clustering, ya que los cl√∫steres son compactos y bien separados.

## F√≥rmula del √çndice de Davies-Bouldin

La f√≥rmula general del √≠ndice es la siguiente:

$$
DBI = \frac{1}{k} \sum_{i=1}^{k} \max_{i \neq j} \left( \frac{s_i + s_j}{d_{ij}} \right)
$$

Donde:
- \( k \): N√∫mero de cl√∫steres.
- \( s_i \): La compacidad del cl√∫ster \( i \), es decir, la distancia media entre los puntos dentro del cl√∫ster \( i \) y su centro.
- \( d_{ij} \): La distancia entre los centros de los cl√∫steres \( i \) y \( j \), \( i \neq j \). Esta distancia se utiliza para medir cu√°n separados est√°n los cl√∫steres.

### Definici√≥n de los t√©rminos:

#### Compacidad \( s_i \):
La compacidad de un cl√∫ster \( i \) se define como la distancia promedio entre cada punto en el cl√∫ster y su centroide. Cuanto m√°s peque√±o sea este valor, m√°s compacto ser√° el cl√∫ster.

La f√≥rmula para la compacidad de un cl√∫ster \( i \) es:

$$
s_i = \frac{1}{n_i} \sum_{x_j \in C_i} \| x_j - \mu_i \|
$$

Donde:
- \( n_i \) es el n√∫mero de puntos en el cl√∫ster \( i \).
- \( \mu_i \) es el centroide del cl√∫ster \( i \), calculado como el promedio de los puntos en \( C_i \).
- \( x_j \) es un punto en el cl√∫ster \( C_i \).

#### Distancia entre centros \( d_{ij} \):
La distancia entre los centros de los cl√∫steres \( i \) y \( j \) se calcula como la distancia euclidiana entre sus respectivos centroides:

$$
d_{ij} = \| \mu_i - \mu_j \|
$$

Donde:
- \( \mu_i \) es el centroide del cl√∫ster \( i \).
- \( \mu_j \) es el centroide del cl√∫ster \( j \).

### √çndice de Davies-Bouldin \( DBI \):
El √≠ndice de Davies-Bouldin se calcula como el promedio de los valores m√°ximos de la relaci√≥n entre la compacidad y la distancia para cada par de cl√∫steres \( i \) y \( j \). Cuanto m√°s bajo sea el valor de \( DBI \), mejor ser√° la calidad del clustering, ya que indica que los cl√∫steres son compactos y bien separados.

### Pasos para calcular el √çndice de Davies-Bouldin:

1. **Calcular los centroides de cada cl√∫ster**: Para cada cl√∫ster \( i \), calcular el centroide \( \mu_i \) como el promedio de los puntos en el cl√∫ster.
2. **Calcular la compacidad \( s_i \) de cada cl√∫ster**: Para cada cl√∫ster \( i \), calcular la distancia media entre cada punto y el centroide de ese cl√∫ster.
3. **Calcular la distancia \( d_{ij} \) entre los centroides de los cl√∫steres**: Para cada par de cl√∫steres \( i \) y \( j \), calcular la distancia euclidiana entre los centroides de esos dos cl√∫steres.
4. **Calcular el √çndice de Davies-Bouldin**: Para cada cl√∫ster \( i \), encontrar el valor m√°ximo de la relaci√≥n \( \frac{s_i + s_j}{d_{ij}} \) para \( j \neq i \). Promediar estos valores m√°ximos y dividir entre el n√∫mero total de cl√∫steres \( k \).

### Interpretaci√≥n del √çndice de Davies-Bouldin:
- **Valor bajo del DBI**: Indica que los cl√∫steres son compactos y bien separados. Esto significa que el algoritmo de clustering ha hecho un buen trabajo.
- **Valor alto del DBI**: Indica que los cl√∫steres est√°n mal definidos, solapados o no bien separados.

### Ejemplo

Supongamos que tenemos tres cl√∫steres y los siguientes puntos:

**Cluster A**: \( (1, 2), (2, 1), (1, 1) \)

**Cluster B**: \( (5, 6), (6, 5), (5, 5) \)

**Cluster C**: \( (8, 9), (9, 8), (8, 8) \)

Los pasos ser√≠an:

1. **Calcular los centroides**:

   - Centro A: \( (1.33, 1.33) \)
   - Centro B: \( (5.33, 5.33) \)
   - Centro C: \( (8.33, 8.33) \)

2. **Calcular la compacidad de cada cl√∫ster**:

   - Para A: calcular la distancia media entre los puntos de A y su centroide \( (1.33, 1.33) \).
   - Para B: calcular la distancia media entre los puntos de B y su centroide \( (5.33, 5.33) \).
   - Para C: calcular la distancia media entre los puntos de C y su centroide \( (8.33, 8.33) \).

3. **Calcular la distancia entre centroides**:

   - \( d_{AB} \): distancia entre los centroides de A y B.
   - \( d_{AC} \): distancia entre los centroides de A y C.
   - \( d_{BC} \): distancia entre los centroides de B y C.

4. **Calcular el DBI**: Promediar los valores m√°ximos de \( \frac{s_i + s_j}{d_{ij}} \) para cada cl√∫ster.


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
```

```python
# Definir los puntos de los cl√∫steres
cluster_a = np.array([[1, 2], [2, 1], [1, 1]])
cluster_b = np.array([[5, 6], [6, 5], [5, 5]])
cluster_c = np.array([[8, 9], [9, 8], [8, 8]])

# Unir todos los puntos en un √∫nico array
X = np.vstack([cluster_a, cluster_b, cluster_c])
```

```python
# Realizar el clustering utilizando KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Calcular el √çndice de Davies-Bouldin
db_score = davies_bouldin_score(X, kmeans.labels_)

# Mostrar el √≠ndice
print(f"√çndice de Davies-Bouldin: {db_score}")
```

```python
# Gr√°fico del Coeficiente de Silueta
sample_silhouette_values = silhouette_samples(X, y_kmeans)

plt.figure(figsize=(8, 4))
y_lower = 10
for i in range(optimal_n_clusters):
    ith_cluster_silhouette_values = sample_silhouette_values[y_kmeans == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values)
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

plt.xlabel("Coeficiente de Silueta")
plt.ylabel("Cluster")
plt.title(f"Gr√°fico del Coeficiente de Silueta para {optimal_n_clusters} clusters")
plt.axvline(x=silhouette_avg, color="red", linestyle="--")
plt.show()
```
