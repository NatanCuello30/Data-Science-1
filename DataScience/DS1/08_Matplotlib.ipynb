```python
# Ejercicio en vivo 
# Levanto el archivo
url = 'https://raw.githubusercontent.com/xlisouski/DataCoder/main/pune_1965_to_2002.csv'
df_lluvias = pd.read_csv(url)
df_lluvias.index = df_lluvias['Year']
df_lluvias = df_lluvias.drop('Year', axis='columns')
df_lluvias.head()
```

```python
# Defino los ejes 
x = df_lluvias.index
y = df_lluvias.sum(axis='columns')
```

```python
# Defino objetos a graficar
fig, ax = plt.subplots(figsize=(12, 5))
#Plot es el metodo para armar un grafico de lineas
ax.plot(x, 
        y, 
        label='Precipitaciones acumuladas')
# Lo tuneamos un poco 
ax.set_xlabel('Ao')  				
ax.set_ylabel('Precipitacin acumulada (mm.)')
ax.set_title('Precipitaciones acumuladas segn el ao')
ax.legend() 
ax.set_xlim(df_lluvias.index[0], df_lluvias.index[-1])
```

```python
# Hago un grfico con ms elementos
fig, ax = plt.subplots(figsize=(12, 5))  
ax.plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero')
ax.plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1')
maximo_enero = df_lluvias['Jan'].max()
maximo_febrero =  df_lluvias['Feb'].max()
ax.axhline(maximo_enero, color='red', linestyle='--', alpha=0.5,    linewidth=3, label='Mxima de enero')
ax.axhline(maximo_febrero, color='red', linestyle=':', alpha=0.5, linewidth=3, label='Mxima de febrero')
ax.set_xlabel('Ao')  
ax.set_ylabel('Precipitacin (mm.)') 
ax.set_title('Precipitaciones de enero y febrero') 	
ax.set_xlim(df_lluvias.index[0], df_lluvias.index[-1])
ax.legend() 
```
